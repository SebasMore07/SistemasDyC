## Pregunta Obligatoria

```python
import numpy as np

# Definir el punto s en el eje imaginario (por ejemplo s = j5)
s = 1j * 5

# Definir la función G(s)H(s) sin la ganancia K
def GH(s):
    return (s + 3) / (s * (s + 1) * (s + 2) * (s + 4))

# Evaluación numérica
val = GH(s)
magnitud = abs(val)
fase_deg = np.angle(val, deg=True)

# Ganancia K necesaria para que |KG(s)H(s)| = 1
K_critico = 1 / magnitud

# Mostrar resultados
print(f"Valor en s = j5: {val}")
print(f"Magnitud: {magnitud:.4f}")
print(f"Ángulo (grados): {fase_deg:.2f}°")
print(f"Ganancia crítica K (para |KGH| = 1): {K_critico:.4f}")
```
