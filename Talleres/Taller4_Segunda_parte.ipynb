```python
# Si estás en Google Colab, descomenta y ejecuta esta línea primero:
# !pip install control
!pip install control
import numpy as np
import matplotlib.pyplot as plt
from control import tf, rlocus

# === Función de Transferencia de lazo abierto ===
# KG(s)H(s) = K(s - 3)(s - 5)/(s + 1)(s + 2)

# Numerador: (s - 3)(s - 5) = s^2 - 8s + 15
num = [1, -8, 15]

# Denominador: (s + 1)(s + 2) = s^2 + 3s + 2
den = [1, 3, 2]

# Sistema de lazo abierto (sin K)
G = tf(num, den)

# === Graficar Lugar Geométrico de las Raíces ===
plt.figure(figsize=(10, 6))
rlocus(G, grid=True)
plt.title("Lugar Geométrico de las Raíces - FTLA (6-13)")
plt.xlabel("Parte real")
plt.ylabel("Parte imaginaria")
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.tight_layout()
plt.show()
```

# PUNTO 5.
 NUMERALES PARA OSACILACION, ESTABILIDAD Y GANACIA

```python
# Si estás en Colab, primero instala la librería:
# !pip install control

import numpy as np
import matplotlib.pyplot as plt
from control import tf, evalfr

# === Definición del sistema en lazo abierto ===
# KG(s)H(s) = K(s^2 - 4s + 20)/(s + 2)(s + 4)

# Numerador: s^2 - 4s + 20
num = [1, -4, 20]
# Denominador: (s + 2)(s + 4) = s^2 + 6s + 8
den = [1, 6, 8]

# Función de transferencia en lazo abierto (sin K)
G = tf(num, den)

# === Búsqueda de frecuencia de cruce de fase (fase ≈ -180°) ===
w_vals = np.linspace(0.1, 10, 1000)
fase_objetivo = -180
tolerancia = 1  # grados
omega_critico = None
K_critico = None

for w in w_vals:
    jw = 1j * w
    Gjw = evalfr(G, jw)
    fase = np.angle(Gjw, deg=True)
    mag = abs(Gjw)

    if np.abs(fase + 180) < tolerancia:
        omega_critico = w
        K_critico = 1 / mag
        break

# Mostrar resultados
print("=== RESULTADOS ===")
print(f"Frecuencia de oscilación (ω) en el borde de la estabilidad: {omega_critico:.4f} rad/s")
print(f"Ganancia crítica K: {K_critico:.4f}")

```
