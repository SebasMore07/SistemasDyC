# LGR Y POLO EN EL ORIGEN DE LAS ECUACIONES 12 Y 13

```python
!pip install control
import numpy as np
import matplotlib.pyplot as plt
from control import tf, rlocus, step_response, pzmap
from control.matlab import margin, pole, zero

# Configuración de gráficos
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['axes.grid'] = True

# Función para graficar LGR con polos, ceros y márgenes
def plot_lgr(sys, title):
    plt.figure()
    rlocus(sys)  # Gráfica del LGR
    poles = pole(sys)
    zeros = zero(sys)
    plt.scatter(np.real(poles), np.imag(poles), marker='x', color='r', label='Polos')
    plt.scatter(np.real(zeros), np.imag(zeros), marker='o', color='b', label='Ceros')
    plt.title(f'Lugar Geométrico de las Raíces: {title}')
    plt.legend()
    plt.show()

    # Mapa de polos y ceros
    plt.figure()
    pzmap(sys)
    plt.title(f'Mapa de Polos y Ceros: {title}')
    plt.show()

    # Margen de ganancia y fase
    gm, pm, wg, wp = margin(sys)
    print(f"Para {title}:")
    print(f"- Margen de ganancia: {gm} (K = {gm} para estabilidad crítica)")
    print(f"- Margen de fase: {pm} grados")

# -------------------------------------------------------------------
# Ecuación (12): G_LA(s) = K(s+3) / (s² + 6s + 11)
# -------------------------------------------------------------------
num12 = [1, 3]  # s + 3
den12 = [1, 6, 11]  # s² + 6s + 11
sys12 = tf(num12, den12)

print("\n" + "="*60)
print("Análisis para la Ecuación (12): G_LA(s) = K(s+3)/(s² + 6s + 11)")
print("="*60)
print("Polos del sistema:", pole(sys12))
print("Ceros del sistema:", zero(sys12))
plot_lgr(sys12, "Ecuación (12)")

# -------------------------------------------------------------------
# Ecuación (13): G_LA(s) = K(s+2) / (s³ + 3s² + 8s + 10)
# -------------------------------------------------------------------
num13 = [1, 2]  # s + 2
den13 = [1, 3, 8, 10]  # s³ + 3s² + 8s + 10
sys13 = tf(num13, den13)

print("\n" + "="*60)
print("Análisis para la Ecuación (13): G_LA(s) = K(s+2)/(s³ + 3s² + 8s + 10)")
print("="*60)
print("Polos del sistema:", pole(sys13))
print("Ceros del sistema:", zero(sys13))
plot_lgr(sys13, "Ecuación (13)")

# -------------------------------------------------------------------
# Análisis de estabilidad para K crítico
# -------------------------------------------------------------------
# Para la ecuación (12): Estable para todo K > 0
# Para la ecuación (13): Estable en -5 < K < 14
print("\n" + "="*60)
print("Resumen de Estabilidad:")
print("="*60)
print("Ecuación (12): Estable para todo K > 0 (no cruce con eje imaginario)")
print("Ecuación (13): Estable para -5 < K < 14")
print("  - K crítico = 14 (polos en ±3.56j)")
```

```python
# --------------------------------------------
# Configuración inicial
# --------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from control import tf, rlocus, pzmap, margin, step_response
from control.matlab import pole, zero

# Configuración de gráficos
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['axes.grid'] = True

# --------------------------------------------
# Función para análisis completo de sistemas
# --------------------------------------------
def analyze_system(sys, title):
    # Lugar Geométrico de las Raíces (LGR)
    plt.figure()
    rlocus(sys)
    plt.title(f'LGR: {title}')
    plt.show()

    # Mapa de Polos y Ceros
    plt.figure()
    pzmap(sys, plot=True, grid=True)
    plt.title(f'Polos y Ceros: {title}')
    plt.show()

    # Margen de Ganancia y Fase
    gm, pm, wg, wp = margin(sys)
    print(f"\nAnálisis para {title}:")
    print(f"- Polos: {pole(sys)}")
    print(f"- Ceros: {zero(sys)}")
    print(f"- Margen de Ganancia: {gm:.2f} (K crítico ≈ {gm:.2f})")
    print(f"- Margen de Fase: {pm:.2f}°")

# --------------------------------------------
# Ecuación (12): G(s) = K(s+3)/(s² + 6s + 11)
# --------------------------------------------
num12 = [1, 3]  # s + 3
den12 = [1, 6, 11]  # s² + 6s + 11
sys12 = tf(num12, den12)

# Ecuación (12) con polo en s=0
den12_mod = [1, 6, 11, 0]  # s(s² + 6s + 11)
sys12_mod = tf(num12, den12_mod)

# --------------------------------------------
# Ecuación (13): G(s) = K(s+2)/(s³ + 3s² + 8s + 10)
# --------------------------------------------
num13 = [1, 2]  # s + 2
den13 = [1, 3, 8, 10]  # s³ + 3s² + 8s + 10
sys13 = tf(num13, den13)

# Ecuación (13) con polo en s=0
den13_mod = [1, 3, 8, 10, 0]  # s(s³ + 3s² + 8s + 10)
sys13_mod = tf(num13, den13_mod)

# --------------------------------------------
# Análisis para todos los sistemas
# --------------------------------------------
print("="*60)
print("ANÁLISIS COMPLETO DE LAS ECUACIONES (12) Y (13)")
print("="*60)

# Ecuación 12 Original
print("\n" + "="*40)
analyze_system(sys12, "Ecuación (12) Original")

# Ecuación 12 con polo en origen
print("\n" + "="*40)
analyze_system(sys12_mod, "Ecuación (12) con Polo en s=0")

# Ecuación 13 Original
print("\n" + "="*40)
analyze_system(sys13, "Ecuación (13) Original")

# Ecuación 13 con polo en origen
print("\n" + "="*40)
analyze_system(sys13_mod, "Ecuación (13) con Polo en s=0")

# --------------------------------------------
# Comparación de estabilidad
# --------------------------------------------
print("\n" + "="*60)
print("RESUMEN DE ESTABILIDAD")
print("="*60)
print("Ecuación (12) Original: Estable para todo K > 0")
print("Ecuación (12) con polo en s=0: Estable para 0 < K < 22")
print("\nEcuación (13) Original: Estable para -5 < K < 14")
print("Ecuación (13) con polo en s=0: Inestable para K grandes (K crítico ≈ 1.5)")

# --------------------------------------------
# Instalación de librería (si es necesario)
# --------------------------------------------
# Si aparece algún error, descomenta y ejecuta:
# !pip install control
```

# LGR Y POLO EN EL ORIGEN PARA LA ECUACION 14 y 15

```python
# --------------------------------------------
# Configuración inicial
# --------------------------------------------
!pip install control --quiet

import numpy as np
import matplotlib.pyplot as plt
from control import tf, rlocus, pzmap, margin, step_response
from control.matlab import pole, zero

# Configuración de gráficos
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['axes.grid'] = True

# --------------------------------------------
# Función para análisis completo de sistemas
# --------------------------------------------
def analyze_system(sys, title):
    # Lugar Geométrico de las Raíces (LGR)
    plt.figure()
    rlocus(sys)
    plt.title(f'LGR: {title}')
    plt.show()

    # Mapa de Polos y Ceros
    plt.figure()
    pzmap(sys, plot=True, grid=True)
    plt.title(f'Polos y Ceros: {title}')
    plt.show()

    # Margen de Ganancia y Fase
    gm, pm, wg, wp = margin(sys)
    print(f"\n🔎 Análisis para {title}:")
    print(f"- Polos: {pole(sys)}")
    print(f"- Ceros: {zero(sys)}")
    print(f"- Margen de Ganancia (K crítico): {gm:.2f}")
    print(f"- Margen de Fase: {pm:.2f}°")

# --------------------------------------------
# SISTEMA 1: Ecuación (14)  GLA(s) = K(s+9)(s−3)/(s⁴ + 10s³ + 70s² + 120s + 250)
# --------------------------------------------
num_14 = [1, 6, -27]  # (s+9)(s-3)
den_14 = [1, 10, 70, 120, 250]
sys_14 = tf(num_14, den_14)

# --------------------------------------------
# SISTEMA 2: Ecuación (15)  GLA(s) = K(s+2)/(s² + 5)
# --------------------------------------------
num_15 = [1, 2]  # (s + 2)
den_15 = [1, 0, 5]  # s^2 + 5
sys_15 = tf(num_15, den_15)

# --------------------------------------------
# ANÁLISIS
# --------------------------------------------
print("="*60)
print("ANÁLISIS COMPLETO DE GLA(s)")
print("="*60)

# Sistema Ecuación (14)
print("\n" + "="*40)
analyze_system(sys_14, "Ecuación (14): K(s+9)(s-3)/(s^4 + 10s^3 + 70s^2 + 120s + 250)")

# Sistema Ecuación (15)
print("\n" + "="*40)
analyze_system(sys_15, "Ecuación (15): K(s+2)/(s^2 + 5)")

# --------------------------------------------
# RESUMEN FINAL
# --------------------------------------------
print("\n" + "="*60)
print("RESUMEN COMPARATIVO")
print("="*60)
print("Ecuación (14): LGR complejo, varios polos, comportamiento oscilatorio e inestabilidad si K > 9.26")
print("Ecuación (15): Segundo orden, subamortiguado según K, margen de estabilidad amplio")
```
