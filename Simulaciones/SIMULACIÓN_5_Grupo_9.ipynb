# FT G1 Y G2 CON UNA ENTRADA u(t) =√2cos(5t)

```python
# RF_1a.py - Simulación del numeral 1.a de SIM_5 (frecuencia ω = 5 rad/s)
# JEAC / Estilo uniforme, Junio 2025

!pip install control
import numpy as n
import matplotlib.pyplot as plt
from control.matlab import *

# Definición de funciones de transferencia
# G1(s) = 30 / (s + 30)
num1 = [30]
den1 = [1, 30]
sys1 = tf(num1, den1)

# G2(s) = 12 / (s^2 + 4s + 8)
num2 = [12]
den2 = [1, 4, 8]
sys2 = tf(num2, den2)

# Frecuencia de interés
w0 = 5  # rad/s

# Rango de frecuencias para análisis
omega = n.logspace(-1, 2, 500)

# Bode plots
plt.figure(1, figsize=(10, 6))
mag1, phase1, omega1 = bode(sys1, omega, plot=False)
plt.subplot(2, 1, 1)
plt.semilogx(omega1, 20 * n.log10(mag1), label="G1(s)")
plt.axvline(w0, color='r', linestyle='--')
plt.ylabel("Magnitud [dB]")
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.semilogx(omega1, n.degrees(phase1), label="G1(s)")
plt.axvline(w0, color='r', linestyle='--')
plt.xlabel("Frecuencia [rad/s]")
plt.ylabel("Fase [°]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Repetir para G2
plt.figure(2, figsize=(10, 6))
mag2, phase2, omega2 = bode(sys2, omega, plot=False)
plt.subplot(2, 1, 1)
plt.semilogx(omega2, 20 * n.log10(mag2), label="G2(s)")
plt.axvline(w0, color='r', linestyle='--')
plt.ylabel("Magnitud [dB]")
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.semilogx(omega2, n.degrees(phase2), label="G2(s)")
plt.axvline(w0, color='r', linestyle='--')
plt.xlabel("Frecuencia [rad/s]")
plt.ylabel("Fase [°]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Encontrar frecuencias de cruce por 0 dB (ganancia unitaria) y -180° para ambos sistemas

def find_crossing(w, y, level):
    """Encuentra la frecuencia donde y cruza el valor level."""
    for i in range(1, len(y)):
        if (y[i-1] - level) * (y[i] - level) < 0:
            # Interpolación lineal
            w0 = w[i-1]
            w1 = w[i]
            y0 = y[i-1]
            y1 = y[i]
            return w0 + (level - y0) * (w1 - w0) / (y1 - y0)
    return None

# Encuentran los cruces
w_gc_10 = find_crossing(w1, mag1, 0)     # Cruce por 0 dB para K=10
w_gc_80 = find_crossing(w2, mag2, 0)     # Cruce por 0 dB para K=80
w_pc = find_crossing(w1, phase1, -180)   # Cruce por -180° (fase común para ambos)

# Graficar
plt.figure(figsize=(12, 8))

# Magnitud
plt.subplot(2, 1, 1)
plt.semilogx(w1, mag1, label="K = 10")
plt.semilogx(w2, mag2, label="K = 80", linestyle='--')
plt.axhline(0, color='grey', linestyle=':', linewidth=1)
if w_gc_10:
    plt.axvline(w_gc_10, color='orange', linestyle=':', label=f'ω₁ (K=10) ≈ {w_gc_10:.2f} rad/s')
if w_gc_80:
    plt.axvline(w_gc_80, color='red', linestyle=':', label=f'ω₁ (K=80) ≈ {w_gc_80:.2f} rad/s')
plt.ylabel("Magnitud (dB)")
plt.title("Diagrama de Bode - Magnitud")
plt.grid(True, which="both", linestyle=":")
plt.legend()

# Fase
plt.subplot(2, 1, 2)
plt.semilogx(w1, phase1, label="K = 10")
plt.semilogx(w2, phase2, label="K = 80", linestyle='--')
plt.axhline(-180, color='red', linestyle=':', linewidth=1)
if w_pc:
    plt.axvline(w_pc, color='purple', linestyle=':', label=f'ω₀ ≈ {w_pc:.2f} rad/s')
plt.ylabel("Fase (°)")
plt.xlabel("Frecuencia (rad/s)")
plt.title("Diagrama de Bode - Fase")
plt.grid(True, which="both", linestyle=":")
plt.legend()

plt.tight_layout()
plt.show()
```
